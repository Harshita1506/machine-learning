```python
import json 
#load the data
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)

    return data 
```

```python
data= load_data("store_data.json")
print(data)
print(type(data))
```

```python
# cleaning and structuring the data 
# having repeated alice DUPLICATED DATA 
# somewhere we have string int float INCONCISTENT # having ratings somewhere 5 and somewhere four  
# user charlie has no age MISSING DATA 
```

```python
# making the data consistent 
# making it clean and structuring the data 
def clean_data(data):

    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5
                }
    cleaned_data=[]
    unique_users=set()

    
    for user in data:
       # print(user["rating"])
        raw_rating=str(user["rating"]).strip().lower()
        # to remove spaces and lower them 
        # problems are that trailing spaces remove 
        # convert text to numbers 
        if raw_rating in text_to_num:
            user["rating"]=text_to_num[user["rating"]]
        #print(user["rating"])
        
        # handling missing values 
        # we will replace misssing data with NaN or if there are more features missing we will remove the entire row 

        raw_age=user.get("age")# .get()function gives value if not present then None 
        if(raw_age==None):
            user["age"]=None

        # to handle duplicate data 
        if(user["name"] in unique_users):
            continue
        unique_users.add(user["name"])  
        cleaned_data.append(user)

    return cleaned_data
        



        
        
```

```python
cleaned_data=(clean_data(data))
```

```python
#getting meaningful insights from the data 
# find the avg rating of a product 
# percentage of users that give poor rating 

def get_insights(data):
    total_rating=0
    for user in data:
        total_rating+=float(user["rating"])# since we had our rating in string 
    print("avg rating is",total_rating/len(data))

    # percentage of users with poor rating 
    poor_rating=0
    for user in data:
        if(float(user["rating"] )<3):
            poor_rating+=1
    print("% of user with poor rating is " ,poor_rating/len(data) *100)
```

```python
get_insights(cleaned_data)
```

```python
# build a reccomendation feature (rule based not ml ) 
# if user rating >= 4, reccoment same brand products 
# id usser rating <4 reccomend different brand 
# we need to create a dict of user and recommended brand 
def get_recomm(data):
    recomm=[]# we will store user info like in dict in a list 
    
    for user in data:
        curr_recom={}
        curr_recom["name"]=user["name"]
        if ((float(user["rating"])) >=4):
            curr_recom["brand"]="apple"
        else:
            curr_recom["brand"]="samsung"
        recomm.append(curr_recom)
    return recomm
            
            
        
        



```
